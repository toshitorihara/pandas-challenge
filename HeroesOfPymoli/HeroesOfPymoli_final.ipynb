```python
# Dependencies and Setup
import pandas as pd

# File to Load
path = "Resources/purchase_data.csv"

# Read and Store into Pandas
data = pd.read_csv(path)
data.head()
```

## Player Count


```python
# Calculate the total number of unique players
total_players = len(data["SN"].value_counts())
player_count = pd.DataFrame({"Total Players":[total_players]})
player_count
```

## Purchasing Analysis (Total)

```python
# Run basic calculations
unique_item = len((data["Item ID"]).unique())
average_price = (data["Price"]).mean()
total_purchase = (data["Purchase ID"]).count()
total_revenue = (data["Price"]).sum()

# Create summary data frame
summary_df = pd.DataFrame({"Number of Unique Items": [unique_item], "Average Price": [average_price], "Number of Purchases": [total_purchase], "Total Revenue": [total_revenue]})

# Formatting
summary_df.style.format({"Average Price": "${:,.2f}", "Total Revenue": "${:,.2f}"})
```

## Age Demographics

```python
# Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins
data["Age Group"] = pd.cut(data["Age"], bins, labels=group_labels)
data

# Calculate the numbers and percentages by age group
age_group = data.groupby("Age Group")
age_count = age_group["SN"].nunique()
age_percent = (age_count / total_players) * 100

# Create summary data frame
age_demographics = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_percent})

# Formatting
age_demographics.index.name = None
age_demographics.style.format({"Percentage of Players":"{:,.2f}%"})
```

## Top Spenders

```python
# Run basic calculations
spender_data = data.groupby("SN")
purchase_count_spender = spender_data["Purchase ID"].count()
avg_purchase_spender = spender_data["Price"].mean()
total_purchase_spender = spender_data["Price"].sum()

# Create summary data frame
top_spenders = pd.DataFrame({"Purchase Count": purchase_count_spender,
                             "Average Purchase Price": avg_purchase_spender,
                             "Total Purchase Value": total_purchase_spender})

# Formatting
summary_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()
summary_spenders.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```
